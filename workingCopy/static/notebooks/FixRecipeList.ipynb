```python
import os
import json
```

```python
file = 'fix_AllRecipes.json'
filename = os.path.join('..', 'data', file)

new_name = 'fix_' + file
fixed_filename = os.path.join('..', 'data', new_name)
```

```python
def GetDefault():
    return {'label': 'Replaced',
            'quantity': 0.0,
            'unit': 'g'}
```

```python
foods = []

with open(filename) as recipe_file:
    foods = json.load(recipe_file)
    
    for food in foods:
        recipe_name = food['recipe']['label']
        total_nutrients = food['recipe']['totalNutrients']
        
        try:
            Calories = total_nutrients['ENERC_KCAL']['quantity']
        except:
            print(f"{recipe_name} Calories not found")
            food['recipe']['totalNutrients']['ENERC_KCAL'] = GetDefault()
        
        try:
            Fat  = total_nutrients['FAT']['quantity']
        except:
            print(f"{recipe_name} Fat not found")
            food['recipe']['totalNutrients']['FAT'] = GetDefault()
            
        try:            
            Sat_Fat = total_nutrients['FASAT']['quantity']
        except:
            print(f"{recipe_name} Sat_Fat not found")
            food['recipe']['totalNutrients']['FASAT'] = GetDefault()
            
        try:
            Trans_Fat = total_nutrients['FATRN']['quantity']
        except:
            print(f"{recipe_name} Trans_Fat not found")
            food['recipe']['totalNutrients']['FATRN'] = GetDefault()

        try:
            Mono_Fat = total_nutrients['FAMS']['quantity']
        except:
            print(f"{recipe_name} Mono_Fat not found")
            food['recipe']['totalNutrients']['FAMS'] = GetDefault()
            
        try:
            Poly_Fat = total_nutrients['FAPU']['quantity']
        except:            
            print(f"{recipe_name} Poly_Fat not found")
            food['recipe']['totalNutrients']['FAPU'] = GetDefault()
            
        try:
            Carbs = total_nutrients['CHOCDF']['quantity']
        except:
            print(f"{recipe_name} Carbs not found")
            food['recipe']['totalNutrients']['CHOCDF'] = GetDefault()
            
        try:
            Fiber = total_nutrients['FIBTG']['quantity']
        except:
            print(f"{recipe_name} Fiber not found")
            food['recipe']['totalNutrients']['FIBTG'] = GetDefault()
            
        try:
            Sugar = total_nutrients['SUGAR']['quantity']
        except:
            print(f"{recipe_name} Sugar not found")
            food['recipe']['totalNutrients']['SUGAR'] = GetDefault()
           
        try:
            Protein = total_nutrients['PROCNT']['quantity']
        except:
            print(f"{recipe_name} Protein not found")
            food['recipe']['totalNutrients']['PROCNT'] = GetDefault()
            
        try:
            Cholesterol = total_nutrients['CHOLE']['quantity']
        except:
            print(f"{recipe_name} Cholesterol not found")
            food['recipe']['totalNutrients']['CHOLE'] = GetDefault()
            
        try:
            Sodium = total_nutrients['NA']['quantity']
        except:
            print(f"{recipe_name} Sodium not found")
            food['recipe']['totalNutrients']['NA'] = GetDefault()
            
        try:
            Calcium = total_nutrients['CA']['quantity']
        except:
            print(f"{recipe_name} Calcium not found")
            food['recipe']['totalNutrients']['CA'] = GetDefault()
            
        try:
            Magnesium = total_nutrients['MG']['quantity']
        except:
            print(f"{recipe_name} Magnesium not found")
            food['recipe']['totalNutrients']['MG'] = GetDefault()
            
        try:
            Potassium = total_nutrients['K']['quantity']
        except:
            print(f"{recipe_name} Potassium not found")
            food['recipe']['totalNutrients']['K'] = GetDefault()
            
        try:
            Iron = total_nutrients['FE']['quantity']
        except:
            print(f"{recipe_name} Iron not found")
            food['recipe']['totalNutrients']['FE'] = GetDefault()
            
        try:
            Zinc = total_nutrients['ZN']['quantity']
        except:
            print(f"{recipe_name} Zinc not found")
            food['recipe']['totalNutrients']['ZN'] = GetDefault()
            
        try:
            Phosphorus= total_nutrients['P']['quantity']
        except:
            print(f"{recipe_name} Phosphorus not found")
            food['recipe']['totalNutrients']['P'] = GetDefault()
            
        try:
            Vit_A= total_nutrients['VITA_RAE']['quantity']
        except:
            print(f"{recipe_name} Vit_A not found")
            food['recipe']['totalNutrients']['VITA_RAE'] = GetDefault()
        
        try:
            Vit_C = total_nutrients['VITC']['quantity']
        except:
            print(f"{recipe_name} Vit_C not found")
            food['recipe']['totalNutrients']['VIT_C'] = GetDefault()
            
        try:
            B1 = total_nutrients['THIA']['quantity']
        except:
            print(f"{recipe_name} B1 not found")            
            food['recipe']['totalNutrients']['B1'] = GetDefault()
            
        try:
            B2 = total_nutrients['RIBF']['quantity']
        except:
            print(f"{recipe_name} B2 not found")
            food['recipe']['totalNutrients']['RIBF'] = GetDefault()
            
        try:            
            B3 = total_nutrients['NIA']['quantity']
        except:
            print(f"{recipe_name} B3 not found")
            food['recipe']['totalNutrients']['NIA'] = GetDefault()
            
        try:            
            B6 = total_nutrients['VITB6A']['quantity']
        except:
            print(f"{recipe_name} B6 not found")
            food['recipe']['totalNutrients']['VITB6A'] = GetDefault()
            
        try:
            Folate_eq = total_nutrients['FOLDFE']['quantity']
        except:
            print(f"{recipe_name} Folate_eq not found")
            food['recipe']['totalNutrients']['FOLDFE'] = GetDefault()
            
        try:
            Folate_food = total_nutrients['FOLFD']['quantity']
        except:
            print(f"{recipe_name} Folate_food not found")
            food['recipe']['totalNutrients']['FOLFD'] = GetDefault()
            
        try:
            B12 = total_nutrients['VITB12']['quantity']
        except:
            print(f"{recipe_name} B12 not found")
            food['recipe']['totalNutrients']['VITB12'] = GetDefault()
            
        try:            
            Vit_D = total_nutrients['VITD']['quantity']
        except:
            print(f"{recipe_name} Vit_D not found")
            food['recipe']['totalNutrients']['VITD'] = GetDefault()
            
        try:            
            Vit_E = total_nutrients['TOCPHA']['quantity']
        except:
            print(f"{recipe_name} Vit_E not found")
            food['recipe']['totalNutrients']['TOCPHA'] = GetDefault()
            
        try:
            Vit_K = total_nutrients['VITK1']['quantity']
        except:
            print(f"{recipe_name} Vit_K not found")
            food['recipe']['totalNutrients']['VITK1'] = GetDefault()
            
        try:
            Water = total_nutrients['WATER']['quantity']             
        except:
            print(f"{recipe_name} Water not found")
            food['recipe']['totalNutrients']['WATER'] = GetDefault()
    
print(json.dumps(foods, sort_keys=True, indent=4))

    
```
